```python
import face_recognition
import cv2
import os
from google.colab.patches import cv2_imshow
def read_img(path):
  img=cv2.imread(path)
  (h, w) = img.shape[:2]
  width=500
  ratio=width/ float(w)
  height= int(h*ratio)
  return cv2.resize(img, (width,height))


Q_encodings=[]
Q_names=[]
Q_directory='Q'

for file in os.listdir(Q_directory):
  img=read_img(Q_directory + '/' + file)
  img_enc= face_recognition.face_encodings(img)[0]
  Q_encodings.append(img_enc)
  Q_names.append(file.split(' - ')[0])
 # print(Q_names)

  K_directory='K'

for file in os.listdir(K_directory):
    print("Processing",file)
    img = read_img(K_directory+'/' +file)
    img_enc=face_recognition.face_encodings(img)[0]
    results=face_recognition.compare_faces(Q_encodings,img_enc)

    aww=face_recognition.face_distance(Q_encodings,img_enc)
    print(aww)




    for i in range(len(results)):
        if(results[i]):
          names=Q_names[i]

          (top,right,bottom,left)=face_recognition.face_locations(img)[0] #get face locations
          cv2.rectangle(img,(left,top), (right,bottom), (0,255,0), 2)
          cv2.putText(img, names, (left + 20, bottom + 20), cv2.FONT_HERSHEY_PLAIN, 1, (255, 255, 255), 1)

          cv2_imshow(img)





```
